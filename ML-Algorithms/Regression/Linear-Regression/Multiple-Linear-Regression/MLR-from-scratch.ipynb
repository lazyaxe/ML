```python
import numpy as np
import pandas as pd

#for comparisions
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
class MyLinearRegression():
    def __init__(self) -> None:
        self.coef_ = np.array([])
        self.intercept_ = 0
        self.weight_matrix = np.array([])

    def fit(self, X, y):
        X = np.asarray(X)
        y = np.array(y)

        ones = np.ones((X.shape[0], 1))
        X = np.hstack((ones, X))

        #pseudo inverse because a real inverse can fail if a matrix is not singular(n x n)
        self.weight_matrix = np.linalg.pinv(X.T @ X) @ X.T @ y

        self.intercept_ = self.weight_matrix[0]
        self.coef_ = self.weight_matrix[1:]

    def predict(self, X):
        X = np.asarray(X)

        ones = np.ones((X.shape[0], 1))
        X = np.hstack((ones, X))

        return X @ self.weight_matrix

    def r2_score(self, y_true, y_pred):
        SSM = np.sum((y_true - np.mean(y_true)) ** 2)
        SSR = np.sum((y_true - y_pred) ** 2)
        R2 = 1 - SSR/SSM
        return R2

    def adjusted_R2(self, X, y):
        R2 = self.r2_score(X, y)
        n = y.lenght
        k = 1
        ratio_numerator = (1 - R2) * (n-1)
        ratio_denominator = n - 1 - k
        adj_R2 = 1 - ratio_numerator/ratio_denominator
        return adj_R2
```

```python
X, y = load_diabetes(return_X_y = True)

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
mlr1 = MyLinearRegression()

mlr1.fit(X_train, y_train)

mlr1.weight_matrix
```

```python

print("My model r2 score", mlr1.r2_score(y_true = y_test, y_pred = mlr1.predict(X_test)))
```

```python
print("My model r2 score", r2_score(y_true = y_test, y_pred = mlr1.predict(X_test)))
```

```python
mlr2 = LinearRegression()
mlr2.fit(X_train, y_train)

print(f"The intercept_ and weight_matrix {mlr2.intercept_} {mlr2.coef_}")
```

```python

print("My model r2 score", r2_score(y_test, mlr2.predict(X_test)))
```
