```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes

#for comparisions
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score
```

### Formula for linear regression:

### $\hat{Y} = X {\beta}$

#### Where,
#### ${\beta} = ({X}^T {X})^{-1} {X}^T Y $

${\beta}$ = The vector of weight/parameters/features size $m$ x 1 <br>
$m$  = number of features

My implementation of the Multiple Linear Regression:

```python
class MyLinearRegression():
    def __init__(self) -> None:
        self.coef_ = np.array([])
        self.intercept_ = 0
        self.weight_vector = np.array([])

    def fit(self, X, y):
        #X is the dataset and essentially a 2D array
        #y is the response variable

        X = np.asarray(X)
        y = np.array(y)

        #We need to add a psuedo-column to the dataset, X
        #because we are calculating the coeffeicients(slope) along with the intercept(B0)
        #and the intercept is not a coefficient of any column in the Dataset X 
        #so we need to add a column of 1s such that the matrix multiplication gives the formula of y_pred,
        #y_pred = X * Weights
        ones = np.ones((X.shape[0], 1))
        X = np.hstack((ones, X))

        #Now we implement the formula using NumPy, Weights = (X_transpose * X)^-1 X_transpose * Y
        #pseudo inverse (np.linalg.pinv) because a regular inverse can fail if a matrix is not singular(n x n)
        self.weight_vector = np.linalg.pinv(X.T @ X) @ X.T @ y

        #Because the intercept was the first element of the Beta/Weight Vector
        self.intercept_ = self.weight_vector[0]
        self.coef_ = self.weight_vector[1:]

    def predict(self, X):
        X = np.asarray(X)

        ones = np.ones((X.shape[0], 1))
        X = np.hstack((ones, X))

        return X @ self.weight_vector
```

My implementation of various measures of accuracy

```python
def my_r2_score( y_true, y_pred):
    SSM = np.sum((y_true - np.mean(y_true)) ** 2)
    SSR = np.sum((y_true - y_pred) ** 2)
    R2 = 1 - SSR/SSM
    return R2

def adjusted_R2( X, y):
    R2 = my_r2_score(X, y)
    n = y.lenght
    k = 1
    ratio_numerator = (1 - R2) * (n-1)
    ratio_denominator = n - 1 - k
    adj_R2 = 1 - ratio_numerator/ratio_denominator
    return adj_R2

def my_mean_absolute_error(y_true, y_pred):
    absolute_error = np.abs(y_true - y_pred)
    return np.sum(absolute_error)/absolute_error.shape[0]

def my_mean_squared_error(y_true, y_pred):
    squared_error = np.square(y_true - y_pred)
    return np.sum(squared_error)/squared_error.shape[0]

def my_root_mean_squared_error(y_true, y_pred):
    return np.sqrt(my_mean_absolute_error(y_true, y_pred))
```

```python
X, y = load_diabetes(return_X_y = True)

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
my_mlr = MyLinearRegression()

my_mlr.fit(X_train, y_train)

print("My Implementation parameters")
print(f"The intercept_ = {my_mlr.intercept_}\n weight_matrix = {my_mlr.coef_}")
```

```python
sk_mlr = LinearRegression()
sk_mlr.fit(X_train, y_train)

print("Sklearn Implementation parameters")
print(f"The intercept_ = {sk_mlr.intercept_}\n weight_matrix = {sk_mlr.coef_}")
```

```python
print("My model mean_absolute_error", mean_absolute_error(y_true = y_test, y_pred = my_mlr.predict(X_test)))
print("Sklearn model mean_absolute_error", mean_absolute_error(y_true = y_test, y_pred = sk_mlr.predict(X_test)))
```

```python
print("My model mean_squared_error", mean_squared_error(y_true = y_test, y_pred = my_mlr.predict(X_test)))
print("Sklearn model mean_squared_error", mean_squared_error(y_true = y_test, y_pred = sk_mlr.predict(X_test)))
```

```python
print("My model root_mean_squared_error", root_mean_squared_error(y_true = y_test, y_pred = my_mlr.predict(X_test)))
print("Sklearn model root_mean_squared_error", root_mean_squared_error(y_true = y_test, y_pred = sk_mlr.predict(X_test)))
```

```python
print("My model r2 score", my_r2_score(y_true = y_test, y_pred = my_mlr.predict(X_test)))
print("Sklearn model r2 score", r2_score(y_true = y_test, y_pred = sk_mlr.predict(X_test)))
```
