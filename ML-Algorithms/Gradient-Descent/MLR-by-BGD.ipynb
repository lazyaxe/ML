Making my own Batch-Gradient-Descent(BGD) Regressor:
* BGD is a type of Gradient Descent Alogrithmn where the BGD uses data of all the rows for a single coefficient's partial derivative and then approximates it.
* Because it uses all the records for approximation of single coefficient, it's only preffered when there are less features(columns) and less rows... 

The formula of the gradient descent: <br>
#### ${\beta}_{new} = {\beta}_{current} - {\eta} {\nabla}_{E}$
where, <br>
* ${\beta}_{current}$ = The currrent value of the coefficient vector
* ${\beta}_{new}$ = new value of the coefficient vector
* ${\eta} $ = learning rate of the gradient descent
* ${\nabla}_{E}$ = The gradient of the error function, $E$

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_diabetes

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
X, y = load_diabetes(return_X_y = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
print(f"Shape of feature {X.shape}, shape of the response {y.shape}")
```

Implementing my own Batch Gradient Regressor, MyBGDRegrssor.

```python
class MyBGDRegressor:
    def __init__(self, epochs: int, learning_rate: float) -> None:
        self.epochs = epochs
        self.lr = learning_rate
        self.coef_ = np.array([])
        self.intercept_ = 0
    
    def fit(self, X_train, y_train):
        #because we also need to calculate the intercept(B0), need a (n * m+1) matrix, where m = number of features
        X_train = np.insert(X_train, 0, 1, axis = 1)
        
        rows, features = X_train.shape
        y_train = y_train.reshape(-1, 1)

        beta_vector = np.zeros(features).reshape(-1, 1)  

        for _ in range(self.epochs):
            y_pred = X_train @ beta_vector

            error = y_pred - y_train.reshape(-1,1)

            gradient = (-2/rows) * X_train.T @ error
            
            beta_vector = beta_vector - self.lr * gradient

        self.intercept_ = beta_vector[0]
        self.coef_ = beta_vector[1 : ]
```

```python
my_bgdr = MyBGDRegressor(epochs = 1000, learning_rate = 0.5)
my_bgdr.fit(X_train = X_train, y_train = y_train)
print(f"The intercept: {my_bgdr.intercept_}, Coefficents: {my_bgdr.coef_}")
```
