```python
import numpy as np
import pandas as pd

#for comparisions
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

Formula for linear regression:

$\hat{Y}_i = \hat{\beta}_0 + \hat{\beta}_1 X_i + \hat{\epsilon}_i$

Where, 
$\hat{\beta}_0$ = Interscept of a line, 
$\hat{\beta}_1$ = Slope of the line, Co-efficent of feature/predictor/independent variable X
Linear Regression's Parameters: 
1. MAE (Mean Absolute Error)
2. MSE (Mean Squared Error)
3. RMSE (Root Mean Squared Error)
4. $R^2$ Score
5. Adjusted $R^2$ Score

```python
class SLR:
    def __init__(self):
        self.b1 = 0
        self.b0 = 0

    def fit_line(self, X_train, y_train):
       #self.b1 = np.cov(X_train, Y_train)/np.var(X_train)

        #or

        x_bar = X_train.mean()
        y_bar = y_train.mean()
        num = 0
        deno = 0
        
        for i in range(X_train.shape[0]):
            num = num + ((X_train[i] - x_bar) * (y_train[i] - x_bar))
            deno = deno + ((X_train[i] - x_bar) * (X_train[i] - x_bar))
        
        self.b1 = num / deno
        self.b0 =  y_bar - (self.b1 * x_bar)
        
        print(f"Intersept = B0 = {self.b0}")
        print(f"Slope = B1 = {self.b1}")

    def calculate_R2(self, X, Y):
        SSM = Y.var()
        SSR = (X.var() - Y) ** 2
        ratio =  SSR/SSM
        R2 = 1 - ratio
        print(f"SSR = {SSR} \n SSM = {SSM}")
        return R2
    
    def adjusted_R2(self, X, Y):
        R2 = self.calculate_R2(X, Y)
        n = Y.shape[0]
        k = n - 1
        ratio_numerator = (1 - R2) * (n-1)
        ratio_denominator = n - 1 - k
        adj_R2 = 1 - ratio_numerator/ratio_denominator
        return adj_R2

    def predict(self, X_test):
        return self.b0 + self.b1 * X_test
```
