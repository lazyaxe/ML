```python
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
class KNN:
    def __init__(self, k = 5) -> None:
        self.n_neighbours = k

    def fit(self, X, y):
        self.X_train = np.asarray(X)
        self.y_train = np.asarray(y)

    def euclidean_distances(self, x, y):
        x = np.asarray(x)
        y = np.asarray(y)
        return np.sqrt(np.dot(x - y, x - y))

    def predict(self, X):
        X = np.asarray(X)
        y_pred = np.zeros(shape=(X.shape[0]))

        for i in range(X.shape[0]):
            #because a new datapoint will have all the previous datapoints as its neighnours
            neighbours = np.zeros(shape=self.X_train.shape[0])

            for j in range(self.X_train.shape[0]):
                neighbours[j] = self.euclidean_distances(X[i], self.X_train[j])

            #After calculating all the distantes, we find the index of K nearest neigbours of the new datapoint based on the distance
            idx = neighbours.flatten().argsort()[: self.n_neighbours]
            nearest_neighbours_classes = []

            for index in idx:
                nearest_neighbours_classes.append(self.y_train[index])

            y_class = Counter(nearest_neighbours_classes).most_common(n=1)

            y_pred[i] = y_class[0][0]

        return y_pred
```

```python
X, y = load_iris(return_X_y = True, as_frame=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=13, test_size=0.2)
```

```python
clf = KNN(k=3)
clf.fit(X_train, y_train)
clf.predict(X_test)
```

```python
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train, y_train)
clf.predict(X_test)
```
